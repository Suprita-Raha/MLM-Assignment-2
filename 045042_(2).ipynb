#### 3.1.6. Data Bifurcation: Training & Testing Sets

  Stratified Sampling: Stratified sampling involves dividing the dataset into homogeneous subgroups, or strata, based on the values of a particular variable. Each stratum represents a distinct category of the variable. In my case, the stratification is based on an the outcome variable 'cluster'.

  Implementation Steps:
  Stratification: The dataset is partitioned into strata based on the values of the outcome variable or feature. Each stratum contains similar instances of the outcome variable.

  Sampling: Within each stratum, samples are randomly selected to form the training and testing sets. The proportion of data allocated to each set is determined by the specified percentages.

  Training Set: 75% of the data is randomly sampled from each stratum and combined to form the training set. This ensures that the training set retains the same proportion of instances from each stratum as the original dataset.

  Testing Set: The remaining 25% of the data is used to form the testing set. Similar to the training set, this subset maintains the proportional representation of instances from each stratum.

Decision Tree:: A decision tree is a popular and intuitive supervised learning algorithm used for both classification and regression tasks. It builds a tree-like structure where each internal node represents a decision based on an input feature, and each leaf node represents a class label or regression value.

Gini Coefficient (or Gini Impurity): In the context of decision trees, Gini impurity is a measure of the impurity or disorder of a set of data points. It quantifies the probability of incorrectly classifying a randomly chosen element if it were randomly labeled according to the distribution of class labels in the set. The Gini coefficient is calculated as the sum of the squared probabilities of each class label being chosen, subtracted from one.

# **5. Managerial Insights**

**5.1. Appropriate Model: Compare and Contrast (Decision Tree | Logistic Regression | Support Vector Machine | K Nearest Neighbour)**

Each of the four models, Decision Tree, Logistic Regression, Support Vector Machine (SVM), and K Nearest Neighbors (KNN), presents distinct characteristics and trade-offs, making them suitable for different scenarios.

The Decision Tree model is intuitive and easy to interpret, making it an excellent choice when transparency and comprehensibility are crucial. It recursively partitions the feature space into segments, allowing for straightforward decision-making. Additionally, it performs well with categorical and numerical data and is robust to outliers. However, Decision Trees may suffer from overfitting, especially with complex datasets, and may not generalize well to unseen data.

Logistic Regression is a linear model widely used for binary classification tasks. It estimates the probability of an instance belonging to a particular class, making it particularly useful for scenarios where understanding the probability of outcomes is essential. It is computationally efficient, handles large datasets well, and is less prone to overfitting compared to Decision Trees. However, Logistic Regression assumes a linear relationship between features and the log-odds of the target variable, which may not hold true in all cases.

Support Vector Machine (SVM) is a powerful algorithm known for its ability to handle high-dimensional data and complex decision boundaries. It works by finding the optimal hyperplane that separates classes in the feature space, maximizing the margin between classes. SVM is effective in cases where the data is not linearly separable, as it can use kernel functions to transform the feature space into a higher dimension where separation is possible. However, SVM can be computationally expensive, especially with large datasets, and the choice of kernel and regularization parameters can impact its performance.

K Nearest Neighbors (KNN) is a non-parametric algorithm that classifies instances based on the majority class among its k nearest neighbors in the feature space. It is simple to understand and implement, making it suitable for quick prototyping and exploratory data analysis. KNN performs well with small to medium-sized datasets and is robust to noisy data. However, it can be computationally expensive during inference, especially with large datasets, as it requires calculating distances between the query instance and all training instances.

*   The Decision Tree model emerges as the most suitable choice for our task based on several key factors. Firstly, its exceptional performance metrics, including perfect accuracy, recall, precision, and F1-score, underscore its robustness in accurately classifying instances. Additionally, the minimal memory usage and negligible training time of the Decision Tree model make it highly efficient, ensuring rapid deployment and low computational overhead. Furthermore, its inherent interpretability allows for clear insights into the decision-making process, enabling stakeholders to understand and trust the model's predictions.

```python
import pandas as pd
import time
import psutil
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Function to measure memory usage
def memory_usage():
    process = psutil.Process()
    return process.memory_info().rss / 1024 ** 2  # Memory usage in MB

# Data preprocessing and splitting
# Assuming you have your data loaded into cars_inputs and cars_output
cars_inputs_train, cars_inputs_test, cars_output_train, cars_output_test = train_test_split(cars_inputs, cars_output, test_size=0.2, random_state=42)

# Results storage
results = []

# Decision Tree
dt_start_time = time.time()
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(cars_inputs_train, cars_output_train)
dt_training_time = time.time() - dt_start_time
dt_memory_used = memory_usage()
dt_pred = dt_model.predict(cars_inputs_test)
dt_accuracy = accuracy_score(cars_output_test, dt_pred)
results.append({'Model': 'Decision Tree', 'Time Taken (s)': dt_training_time, 'Memory Used (MB)': dt_memory_used, 'Accuracy': dt_accuracy})

# KNN
k_values = [5, 7, 9, 11, 13, 15]
for k in k_values:
    knn_start_time = time.time()
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(cars_inputs_train, cars_output_train)
    knn_training_time = time.time() - knn_start_time
    knn_memory_used = memory_usage()
    knn_pred = knn.predict(cars_inputs_test)
    knn_accuracy = accuracy_score(cars_output_test, knn_pred)
    results.append({'Model': f'KNN (k={k})', 'Time Taken (s)': knn_training_time, 'Memory Used (MB)': knn_memory_used, 'Accuracy': knn_accuracy})

# Logistic Regression
lr_start_time = time.time()
logreg = LogisticRegression(random_state=45042, solver='liblinear')
logreg.fit(cars_inputs_train, cars_output_train)
lr_training_time = time.time() - lr_start_time
lr_memory_used = memory_usage()
lr_pred = logreg.predict(cars_inputs_test)
lr_accuracy = accuracy_score(cars_output_test, lr_pred)
results.append({'Model': 'Logistic Regression', 'Time Taken (s)': lr_training_time, 'Memory Used (MB)': lr_memory_used, 'Accuracy': lr_accuracy})

# SVM
svm_start_time = time.time()
svm_classifier = SVC(kernel='rbf', random_state=45042)
svm_classifier.fit(cars_inputs_train, cars_output_train)
svm_training_time = time.time() - svm_start_time
svm_memory_used = memory_usage()
svm_pred = svm_classifier.predict(cars_inputs_test)
svm_accuracy = accuracy_score(cars_output_test, svm_pred)
results.append({'Model': 'SVM', 'Time Taken (s)': svm_training_time, 'Memory Used (MB)': svm_memory_used, 'Accuracy': svm_accuracy})

# Create DataFrame
results_df = pd.DataFrame(results)

# Display the results
print(results_df)
```
